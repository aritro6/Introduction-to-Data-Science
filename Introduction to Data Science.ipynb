## Hypothesis Testing:

```python
#Taken from http://hamelg.blogspot.com/2015/11/python-for-data-analysis-part-24.html

#One Sample T-Test: Test whether the average age of voters Minnesota differs from the population
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math

np.random.seed(6)

population_ages1 = stats.poisson.rvs(loc=18, mu=35, size=150000)
population_ages2 = stats.poisson.rvs(loc=18, mu=10, size=100000)
population_ages = np.concatenate((population_ages1, population_ages2))

minnesota_ages1 = stats.poisson.rvs(loc=18, mu=30, size=30)
minnesota_ages2 = stats.poisson.rvs(loc=18, mu=10, size=20)
minnesota_ages = np.concatenate((minnesota_ages1, minnesota_ages2))

print('Mean Population Ages: ' + str(population_ages.mean()))
print('Mean Minnesota Ages: ' + str(minnesota_ages.mean()))

stats.ttest_1samp(a= minnesota_ages, popmean= population_ages.mean())
```

```python
# Scipy allows for workign with popular probability distributions, as well as hypothesis testing
import scipy.stats as scs
import matplotlib.pyplot as plt

%matplotlib inline

#create the Random Varible X with a mean of 0 and standard deviation of 1
null = scs.norm(loc=0,scale=1)
alt = scs.norm(loc=3,scale=1)

x = np.linspace(-5,8, 100)

plt.plot(x, null.pdf(x), label='Null')
plt.plot(x, alt.pdf(x), label='Alt')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```
